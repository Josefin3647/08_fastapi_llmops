# Prompt creation

```python
from mlflow.genai.prompts import register_prompt

support_agent_prompt = register_prompt(
    name="support_agent_system_prompt",
    template="""You are a customer support agent for EvilCorp.
                Use the tool lookup_faq to get relevant context before
                answering.""",
    tags={
        "author": "josefin",
        "agent": "support_agent",
        "prompt_type": "system_prompt"
    }
)
```

```python
faq_lookup_docstring = register_prompt(
    name="lookup_faq_description",
    template="""Looks up the answer for question based in keyword""",
    tags={
        "author": "josefin",
        "prompt_type": "tool_description"
    }
)
faq_lookup_docstring
```
